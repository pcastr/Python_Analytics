# Adicição de Colunas e Índices com e Sem Broadcasting

### Sem Broadcasting

```python
# Carrega um arquivo do disco e armazena como dataframe
df = pd.read_csv("dados/dataset1.csv")
```

```python
df.head()
```

```python
# Adicionando nova coluna
df['final_salary'] = 0
```

```python
df.head()
```

```python
# Preencendo o valor da coluna a partir de outras variáveis 
df['final_salary'] = df['base_salary'] + df['guaranteed_compensation']; df.head()
```

```python
# Adicionando nova coluna
df.insert(0, column='ID', value = range(1, 1 + len(df))); df.head()
```

### Com Broadcasting

> Broadcasting é propagação de uma operação ao londo do dataframe

```python
# Esta instrução não altera o dataframe original
df['base_salary'].add(5)
```

```python
df.base_salary.head()
```

```python
# Esta instrução altera o dataframe original
df['base_salary'] = df['base_salary'].add(5)
```

```python
df.base_salary.head()
```

```python
# Adicionando nova coluna usando Broadcasting (convertendo USD para EURO, neste exemplo)
df['base_salary_eur'] = df['base_salary'].mul(0.93); df.head()
```

```python
# Drop de uma coluna
df.drop(columns=['base_salary_eur'], inplace = True); df.head()
```

```python
# Adicionando nova coluna 
df.insert(6, column='base_salary_eur', value=df['base_salary'].mul(0.93)); df.head()
```

```python
# Convertemos a coluna ID em índice da tabela 
df.set_index('ID', inplace=True); df.head()
```

```python
# Renomeando as colunas
df.rename(columns= {'base_salary': 'base_salary_usd',
                   'guaranteed_compensation':'guaranteed_compensation_usd',
                   'final_salary':'final_salary_usd'},inplace=True); df.head()
```
